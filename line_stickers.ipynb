```python
import requests as req
from bs4 import BeautifulSoup as bs
import os, json # os 套件提供的各種操作系統功能
from fake_useragent import UserAgent # 生成隨機的使用者代理字串
from pprint import pprint
```

```python
ua = UserAgent()
# print('虛擬 User Agent :', ua.random)

# 建立儲存圖片的資料夾
folderPath = 'line_stickers'
if not os.path.exists(folderPath):
    os.makedirs (folderPath)
    
# 儲存 Line 貼圖的 list
listLinestickers = []

# 自訂標頭
my_headers = {
    'user-agent' : ua.random
}

url = 'https://store.line.me/stickershop/product/1585840/zh-Hant?from=sticker'
# 將自訂標頭加入 get 請求當中
res = req.get(url,headers = my_headers)
# 建立 soup 物件
soup = bs(res.text,'lxml')
# 取得所有 ul 中的 li 元素（放置貼圖的元素）
li_element = soup.select('ul.mdCMN09Ul.FnStickerList > li.mdCMN09Li.FnStickerPreviewItem.animation-sticker ')

# 逐一取得 li 元素中的 data-preview 資訊
for li in li_element:
    # 取得 data-preview 屬性的值
    strJson = li['data-preview']
    # 以 Json 剖析其中的值
    obj = json.loads(strJson)
    # 加入 list 中
    listLinestickers.append(obj)

# 下載 listLinestickers 中的檔案
for obj in listLinestickers:
    os.system(f'curl {obj["animationUrl"]} -o {folderPath}/{obj["id"]}.gif')
```
