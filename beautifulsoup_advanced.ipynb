```python
from bs4 import BeautifulSoup as bs
import requests as req
from pprint import pprint
```

```python
url = 'https://www.ptt.cc/bbs/NBA/index.html'
res = req.get(url)
soup = bs(res.text,'lxml') # 指定 lxml 作為解析器
# 建立 List 放置列表資訊
list_posts = []
```

```python
list_posts.clear()
# 取得文章標題及連結
for x in soup.select('div.r-ent > div.title > a[href]'): # a 中必須包含 href 才要，沒有 href 的可能是被刪文了
    # print(x.get_text())
    # print('https://www.ptt.cc' + x['href'])

    list_posts.append({
        'title' : x.get_text(),
        'link' : 'https://www.ptt.cc' + x['href']
    })
pprint(list_posts)
```

```python
# 分別進入每篇文章並查看
for index, obj in enumerate(list_posts):
    res_ = req.get(obj['link'])
    soup_ = bs(res_.text,'lxml')
    
    # 去除作者、標題、時間等內容
    for div in soup_.select('div[class ^= "article-metaline"]'): # 開頭為 article-metaline
        div.decompose() # 刪去所選內容
        
    # 去除推、→、噓等內容
    if(len(soup_.select('div.push')) > 0):
        for div in soup_.select('div.push'):
            div.decompose()
            
    # 去除發信站、文章網址等內容
    if(len(soup_.select('span.f2')) > 0):
        for span in soup_.select('span.f2'):
            span.decompose()
            
    # 取得實際需要的內容 (類似 JavaScript 的 innerHTML)
    html = soup_.select_one('div#main-content').decode_contents()
    # html = str(soup_.select_one('div#main-content')) # 類似 JavaScript outerHTML

    # 加入到 list_posts 中
    list_posts[index]['content'] = html
    
print(list_posts)
```

# 取得多個分頁內容

```python
list_posts = []
for page in range(1,6501):
    url = f'https://www.ptt.cc/bbs/NBA/index{page}.html'
    res = req.get(url)
    soup = bs(res.text,'lxml')
    for a in soup.select('div.r-ent > div.title > a[href]'):
        list_posts.append({
            'title' : a.get_text(),
            'link' : 'https://www.ptt.cc' + a['href']
        })
pprint(list_posts)
```
